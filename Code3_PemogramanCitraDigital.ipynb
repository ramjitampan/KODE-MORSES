```python
import numpy as np

# Resize (sampling)
sizes = [128, 64, 32]
resized_imgs = [cv2.resize(img, (s, s), interpolation=cv2.INTER_LINEAR) for s in sizes]

# Konversi grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Kuantisasi
def quantize(img_gray, levels):
    step = 256 // levels
    return (img_gray // step) * step

q256 = quantize(gray, 256)
q16  = quantize(gray, 16)
q4   = quantize(gray, 4)

# Tampilkan hasil
fig, ax = plt.subplots(2, 3, figsize=(15, 8))
for i, s in enumerate(sizes):
    ax[0][i].imshow(cv2.cvtColor(resized_imgs[i], cv2.COLOR_BGR2RGB))
    ax[0][i].set_title(f"Resolusi {s}x{s}")
    ax[0][i].axis("off")

ax[1][0].imshow(q256, cmap="gray"); ax[1][0].set_title("Kuantisasi 256 level"); ax[1][0].axis("off")
ax[1][1].imshow(q16, cmap="gray"); ax[1][1].set_title("Kuantisasi 16 level"); ax[1][1].axis("off")
ax[1][2].imshow(q4, cmap="gray"); ax[1][2].set_title("Kuantisasi 4 level"); ax[1][2].axis("off")
plt.show()
```
