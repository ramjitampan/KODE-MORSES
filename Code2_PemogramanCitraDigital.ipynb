```python
import cv2
import matplotlib.pyplot as plt

# Baca citra
img = cv2.imread("/content/RAMZY FOTO.jpg")
h, w, c = img.shape
print("Resolusi:", w, "x", h)
print("Kedalaman warna:", c)

# Konversi warna
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_ycbcr = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Visualisasi
fig, ax = plt.subplots(1, 4, figsize=(18, 5))
ax[0].imshow(img_rgb); ax[0].set_title("RGB"); ax[0].axis("off")
ax[1].imshow(img_hsv[:,:,0], cmap="hsv"); ax[1].set_title("HSV (Hue)"); ax[1].axis("off")
ax[2].imshow(img_ycbcr[:,:,0], cmap="gray"); ax[2].set_title("YCbCr (Y)"); ax[2].axis("off")
ax[3].imshow(gray, cmap="gray"); ax[3].set_title("Grayscale"); ax[3].axis("off")
plt.show()
```
